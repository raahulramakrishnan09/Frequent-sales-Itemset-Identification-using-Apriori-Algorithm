```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df=pd.read_excel('/content/Detailed-Sales-Report.xlsx',header=None)
df
```

```python
df=df.iloc[2:,[2,4]]
df.columns=['INVOICE_NO','ITEM_NAME']
df
```

```python
df['INVOICE_NO'].nunique()
```

```python
grp=df.groupby('INVOICE_NO')['ITEM_NAME'].apply(list).reset_index()
grp
```

```python
df=grp['ITEM_NAME'].apply(pd.Series)
df
```

```python
df.shape
```

```python
transactions=[]
for i in range(0,len(df)):
  transactions.append([str(df.values[i,j]) for j in range(0,9) if str(df.values[i,j])!='0'])
transactions
```

```python
!pip install apyori
```

```python
#apriori_algorithm
from apyori import apriori
rules=apriori(transactions=transactions,min_support=0.0001,min_cinfidence=0.2,min_lift=3,min_length=2,max_length=2)
results=list(rules)
results
```

```python
def inspect(results):
  lhs=[tuple(i[2][0][0])[0] for i in results]
  rhs=[tuple(i[2][0][1])[0] for i in results]
  support=[i[1] for i in results]
  confidence=[i[2][0][2] for i in results]
  lift=[i[2][0][3] for i in results]
  return list(zip(lhs,rhs,support,confidence,lift))
data=pd.DataFrame(inspect(results),columns=['Item1','Item2','Support','Confidence','Lift'])
data
```

```python
#top10_frequent_Itemset
data.nlargest(n=10,columns='Lift')
```
